```python
import pandas as pd
file_mit_test = 'mitbih_test.csv'
file_mit_train = 'mitbih_train.csv'
file_ptbdb_ab = 'ptbdb_abnormal.csv'
file_ptbdb_no = 'ptbdb_normal.csv'

df_mit_test = pd.read_csv(file_mit_test, header=None)
df_mit_train = pd.read_csv(file_mit_train, header=None)
df_ptbdb_ab = pd.read_csv(file_ptbdb_ab, header=None)
df_ptbdb_no = pd.read_csv(file_ptbdb_no, header=None)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# (1) Linear regression
from sklearn import linear_model
X_train = df_mit_train.iloc[:,:-1]
Y_train = df_mit_train.iloc[:,-1]
regr_linear = linear_model.LinearRegression()
regr_linear.fit(X_train,Y_train)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test = df_mit_test.iloc[:,:-1]
Y_test = df_mit_test.iloc[:,-1]
```

```python
print('Linear regression:', regr_linear.score(X_test,Y_test))
```

```python
#(2) lasso regression
lasso = linear_model.Lasso()
lasso.fit(X_train,Y_train)
print('Lasso regression score:', lasso.score(X_test,Y_test))
# Linear regression is better than Lasso. 
```

```python
#(3) Grid Search for Lasso regressor
from sklearn.model_selection import GridSearchCV, cross_val_score

alphas = np.logspace(-3,0,50)
gscv = GridSearchCV(estimator=lasso,param_grid=dict(alpha=alphas),return_train_score=False)
gscv.fit(X_train,Y_train)

# plot results
plt.plot(alphas,gscv.cv_results_['mean_test_score'])
plt.fill_between(alphas,
                gscv.cv_results_['mean_test_score']+gscv.cv_results_['std_test_score'],
                gscv.cv_results_['mean_test_score']-gscv.cv_results_['std_test_score'],
                alpha = 0.1)
plt.xscale('log')
plt.xlabel('$\\alpha$')
plt.ylabel('$R^2$')
```

```python
best_lasso = gscv.best_estimator_

print('Linear regression score:', regr_linear.score(X_test, Y_test))
print('Default Lasso regression score:', lasso.score(X_test, Y_test))
print('Best Lasso regression (alpha={:0.3f}) score:'.format(best_lasso.alpha), best_lasso.score(X_test, Y_test))
```
